```python
from operator import itemgetter, attrgetter

def test_key(x):
    return x.first_prior[x.category]

class Event:
    first_prior = {'partners': 1, 'run': 2, 'lifestyle': 3}
    def __init__(self,subject,category, num):
        self.subject = subject
        self.category = category
        self.num = num
    def __str__(self):
        return "{}-{}-{}".format(self.subject,self.category, self.num)

Event_objects = []


user = 101
df_user = df[df['tabnum'] == user ]

#my = pd.DataFrame(df_user, columns= ['category','participants_count'])
for index, row in df_user.iterrows():
    Event_objects.append(Event(row['subject'],row['category'],row['participants_count']))

    
sort = sorted(Event_objects,key=attrgetter('num') , reverse= True)
s = sorted(sort, key = test_key)


rows = []
for i, a in enumerate(s, start=1):
    rows.append([i, a.subject])
    
df_my = pd.DataFrame(rows, columns=["prior", "subject"])
df_my.head(10)
```
